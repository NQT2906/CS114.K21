# **Import library**

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import cv2
import imageio
import glob
import pickle
from skimage.feature import hog
from skimage import data, exposure
from sklearn.metrics import classification_report
```

# **Split data to train and test**

```
X_train = []
X_test = []
Y_train = []
Y_test = []
```

```
path = []
for im_path in glob.glob('Folio_Test/*'):
  path.append(im_path)
path.sort()
for im_path in path: 
    cv_img = cv2.imread(im_path);     
    gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)   
    img = cv2.resize(gray,(32, 32))   
    X_test.append(img)
Y_test = pd.read_csv('Label_Test/Label_Test.csv')
```

```
path = []
for im_path in glob.glob('Folio_Train/*'):
  path.append(im_path)
path.sort()
for im_path in path: 
    cv_img = cv2.imread(im_path);     
    gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)   
    img = cv2.resize(gray,(32, 32))   
    X_train.append(img)
Y_train = pd.read_csv('Label_Train/Label_Train.csv')
```

**Backup data if you want to get new data and needn't load data**

```
X_Train = X_train
X_Test = X_test
Y_Train = Y_train
Y_Test = Y_test
```

**Run in this if you want to use old data**

```
X_train = X_Train
X_test = X_Test
Y_train = Y_Train
Y_test = Y_Test
```

```
X_train = np.array(X_train)
Y_train = np.array(Y_train)
X_test = np.array(X_test)
Y_test = np.array(Y_test)
```

# **Transform label**

**Transform test label**

```
le = LabelEncoder()
le.fit(Y_test)
Y_test = le.transform(Y_test)
```

**Transform train label**

```
le = LabelEncoder()
le.fit(Y_train)
Y_train = le.transform(Y_train)
```

# **Hog of data train + data test**

```
#For X_train
X_train_feature = []
for i in range(len(X_train)):
    feature = hog(X_train[i], orientations = 10, pixels_per_cell = (2, 2), cells_per_block = (1,1), block_norm = "L2")
    X_train_feature.append(feature)
X_train_feature = np.array(X_train_feature, dtype = np.float32)


#For X_test
X_test_feature = []
for i in range(len(X_test)):
    feature = hog(X_test[i], orientations = 10, pixels_per_cell = (2, 2), cells_per_block = (1,1), block_norm = "L2")
    X_test_feature.append(feature)
X_test_feature = np.array(X_test_feature, dtype=np.float32)
```

# **No hog**

```
SC = StandardScaler()

nsamples, nx, ny = X_train.shape

X_train_shape = X_train.reshape(nsamples, nx * ny)

X_train_shape = SC.fit_transform(X_train_shape)

Y_train_shape = Y_train.reshape(-1, 1)

nsamples, nx, ny = X_test.shape

X_test_shape = X_test.reshape(nsamples, nx * ny)

X_test_shape = SC.transform(X_test_shape)

Y_test_shape = Y_test.reshape(-1, 1)
```

# **Train model**

**Choose SVM model**

```
classifier = SVC(kernel = "linear")
```

**Choose Logistic Regression (My best choose)**

```
classifier = LogisticRegression(random_state = 0, max_iter = 800)
```

**Train with HOG**

```
classifier.fit(X_train_feature, Y_train)

Y_train_pred = classifier.predict(X_train_feature)
Y_test_pred = classifier.predict(X_test_feature)
```

**Train with StandardScaler (My best choose)**

```
classifier.fit(X_train_shape, Y_train)

Y_train_pred = classifier.predict(X_train_shape)
Y_test_pred = classifier.predict(X_test_shape)
```

# **Confusion matrix and accuracy score of data**

**Accuracy score**

```
print("Accuracy score of train:", accuracy_score(Y_train, Y_train_pred))
print("Accuracy score of test:", accuracy_score(Y_test, Y_test_pred))
```

**Heat map data train**

```
score = accuracy_score(Y_train, Y_train_pred)
cm = confusion_matrix(Y_train, Y_train_pred)

plt.figure(figsize = (32, 32))
print('Heat map:')
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

**Heat map data test**

```
score = accuracy_score(Y_test, Y_test_pred)

cm = confusion_matrix(Y_test, Y_test_pred)

plt.figure(figsize = (32, 32))
print('Heat map:')
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```
