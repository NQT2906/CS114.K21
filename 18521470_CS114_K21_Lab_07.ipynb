```
import turicreate
```

# **Load data**

```
# Load data từ git
!git clone "https://github.com/NQT2906/CS114.K21" Dataset
```

```
cd Dataset/Sales
```

```
# Đọc data bằng phương thức SFrame của thư viện turicreate
sales = turicreate.SFrame('home_data.sframe/')
```

```
# Show data
sales
```

```
# Show thông tin của từng thuộc tính trong bộ data
sales.show()
```

# **Split train and test set**

```
# Chia bộ data thành 2 phần train và test ( train chiếm 80%, test chiếm 20%)
training_set, test_set = sales.random_split(.8, seed=0)
```

# **Train simple regression model**

```
# Tạo mô hình hồi quy tuyến tính của thư viện turricreate. Mô hình được gán cho sqft_model.
# Mục tiêu của mô hình là dự đoán giá nhà, thuộc tính dùng để train cho mô hình là sqft_living
sqft_model = turicreate.linear_regression.create(training_set,target='price',features=['sqft_living'])
```

```
# Trung bình giá nhà trong bộ test
print (test_set['price'].mean())
```

```
# Đánh giá mô hình bằng bộ dữ liệu test
print (sqft_model.evaluate(test_set))
```

# **Explore model a little further**

```
sqft_model.coefficients
```

```
# Import pyplot từ thư viện matplotlib và gán cho nó là plt
import matplotlib.pyplot as plt
```

# **Visualize data and model**

```
# Visualize dữ liệu train bằng phương thức plot của plt(*)
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(training_set['sqft_living'],training_set['price'],'.',
        training_set['sqft_living'],sqft_model.predict(training_set),'-')
plt.xlabel('SQFT Living')
plt.ylabel('Price')
plt.title('Price and Living room')
```

```
# Visualize dữ liệu test bằng phương thức plot của plt(*)
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_set['sqft_living'],test_set['price'],'.',
        test_set['sqft_living'],sqft_model.predict(test_set),'-')
plt.xlabel('SQFT Living')
plt.ylabel('Price')
plt.title('Price and Living room')
```

# **Build a model with these additional features**

```
# Tạo mô hình hồi quy tuyến tính đa biến với thuộc tính là my_features còn mục tiêu vẫn là "price"
my_features_model = turicreate.linear_regression.create(training_set,target='price',features=my_features)
```

# **Compare simple model with more complex one**

```
# Đánh giá 2 mô hình bằng bộ test
print (sqft_model.evaluate(test_set))
print (my_features_model.evaluate(test_set))
```
